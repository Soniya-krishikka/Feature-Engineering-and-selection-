```python
# 1. What is the overall distribution of patients with and without heart disease in this dataset?
print("Distribution of Heart Disease (target):")
print(df['target'].value_counts())
print()
```

```python
# plotting bar chart
plt.figure(figsize=(6, 4))
sns.countplot(x='target', data=df, palette='Set2')
plt.title('Distribution of Target Variable (Heart Disease Presence)')
plt.xlabel('Heart Disease')
plt.ylabel('Count')
plt.show()
```

```python
# 2. Which features have the highest correlation with the target variable (heart disease)?
print("Correlation with target:")
print(df.corr()['target'].sort_values(ascending=False))
print()
```

```python
# plotting
plt.figure(figsize=(8, 4))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 3. How does the age distribution differ between patients with and without heart disease?
print("Age distribution by heart disease presence:")
print(df.groupby('target')['age'].describe())
print()
```

```python
# plotting
plt.figure(figsize=(8, 4))
sns.histplot(data=df, x='age', hue='target', multiple='stack', palette='Set2', kde=True)
plt.title('Age Distribution by Heart Disease Presence')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# 4. What are the typical cholesterol levels in patients with heart disease compared to those without it?
print("Cholesterol levels by heart disease presence:")
print(df.groupby('target')['chol'].describe())
print()
```

```python
# plotting
plt.figure(figsize=(8, 4))
sns.boxplot(x='target', y='chol', data=df, palette='Set3')
plt.title('Cholesterol Levels by Heart Disease Presence')
plt.xlabel('Heart Disease')
plt.ylabel('Cholesterol (mg/dl)')
plt.show()
```

```python
# 5. What is the relationship between age and maximum heart rate achieved?
print("Age and max heart rate (thalach) by heart disease presence:")
print(df.groupby('target')[['age', 'thalach']].describe())
print()
```

```python
# plotting 
plt.figure(figsize=(8, 4))
sns.scatterplot(x='age', y='thalach', hue='target', palette='Set2', data=df)
plt.title('Max Heart Rate Achieved vs Age')
plt.xlabel('Age')
plt.ylabel('Max Heart Rate Achieved (thalach)')
plt.show()
```

```python
# 6. Which features are most important for predicting heart disease according to the Random Forest model?
from sklearn.ensemble import RandomForestClassifier

# Split data into features and target
X = df.drop(columns=['target'])
y = df['target']

# Train a RandomForest model to find feature importance
clf = RandomForestClassifier(random_state=42)
clf.fit(X, y)
```

```python
# Get feature importances
importances = clf.feature_importances_
indices = np.argsort(importances)[::-1]
features = X.columns

print("Feature importances:")
for i in range(len(features)):
    print(f"{features[indices[i]]}: {importances[indices[i]]:.4f}")
print()
```

```python
# Plot feature importances
plt.figure(figsize=(8, 4))
sns.barplot(x=importances[indices], y=features[indices], palette='viridis')
plt.title('Feature Importances')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```

```python
# 7. How do pairs of features relate to each other when separated by the presence or absence of heart disease?
print("Pairplot (not shown) will display pairwise relationships among features.")
plt.figure(figsize=(12, 10))
sns.pairplot(df, hue='target', palette='Set1')
plt.suptitle('Pairplot of Features', y=1.02)
plt.show()
```

```python
# 8. Are there any potential interactions between features that might affect heart disease prediction?
print("Interaction between oldpeak and thalach:")
print(df.groupby('target')[['oldpeak', 'thalach']].describe())
print()
```

```python
# plotting
plt.figure(figsize=(8, 4))
sns.scatterplot(x='oldpeak', y='thalach', hue='target', palette='Set2', data=df)
plt.title('Interaction between Oldpeak and Max Heart Rate by Heart Disease Presence')
plt.xlabel('Oldpeak (ST depression)')
plt.ylabel('Max Heart Rate Achieved (thalach)')
plt.show()
```
